1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.
- Descriptive statistics summarize and describe data you already have.
- Examples: mean, median, mode, charts, tables.
- Example: Average marks of a class = 75

- Inferential statistics use a sample to make conclusions about a population.
- Examples: hypothesis testing, confidence intervals, prediction.
- Example: Surveying 100 customers to estimate satisfaction of all customers.

2. What is sampling in statistics? Explain the differences between random and stratified sampling.
- Sampling means selecting a subset of data from a population to represent the whole.

- Random Sampling:
- Every individual has equal chance
- Simple and unbiased
- Example: randomly picking 50 students from a school

- Stratified Sampling:
- Population divided into groups (strata)
- Sample taken from each group
- Example: selecting students from each grade level

3. Define mean, median, and mode. Explain why these measures of central tendency are important.
-Mean = average
- Median = middle value
- Mode = most frequent value
- They summarize data and help understand its center.
- Example: Median is useful when data has outliers.

4. Explain skewness and kurtosis. What does a positive skew imply about the data?
- Skewness: measures asymmetry
- Positive skew → tail on right side (most values are small)
- Kurtosis: measures how peaked or flat the distribution is
- Positive skew means most data is concentrated on the left with a long right tail.

6.  Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
- import numpy as np

  list_x = [10, 20, 30, 40, 50]

  list_y = [15, 25, 35, 45, 60]

  covariance = np.cov(list_x, list_y)[0][1]

  correlation = np.corrcoef(list_x, list_y)[0][1]

  print("Covariance:", covariance)

  print("Correlation:", correlation)

- Output:
Covariance: 312.5
Correlation: 0.994376
